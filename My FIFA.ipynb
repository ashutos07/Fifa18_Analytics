## 3. Data Cleaning

#### edting columns

```python
df_copy= df.copy()
```

```python
list(df_copy)
```

```python
df_copy.rename(columns={'Goal Scored':'Goal_Scored',
'Ball Possession %':'Ball_Possession',
'Free Kicks':'Free_Kicks',
'Pass Accuracy %':'Pass_Accuracy',
'Distance Covered (Kms)':'Distance_Covered',
'Fouls Committed':'Fouls_Committed',
'Yellow Card':'Yellow_Card',
'Yellow & Red':'Yellow & Red',
'Man of the Match':'Man_of_the_Match',
'1st Goal':'1st_Goals',
'Goals in PSO':'Goals_in_PSO',
'Own goals':'Own_goals',
'Own goal Time':'Own_goal_Time'}, inplace=True)
```

```python
df_copy.head()
```

#### dealing with NaN
converting NaN under Own_goals to 0

```python
df_copy['Own_goals']=df_copy['Own_goals'].fillna(0)
```

```python
df_copy
```

#### dealing with Yes and No
coverting 'Yes' and 'No' values of 'Man_of_the_Match' and 'PSO' to numeric values

```python
df_copy['Man_of_the_Match']= df_copy['Man_of_the_Match'].map({'Yes':1,'No':0})
```

```python
df_copy['PSO']=df_copy['PSO'].map({'Yes':1,'No':0})
```

```python
df_copy.head()
```

#### dealing with unnecessary columns
removing columns not required for this analysis

```python
columns=['Own_goal_Time','1st_Goals']
df_copy.drop(columns,axis=1,inplace=True)
```

```python
df_copy.info()
```

#### dealing with mismatch
converting datatypes 

```python
df_copy['Own_goals']=df_copy['Own_goals'].astype('int64')
```

```python
df_copy.info()
```

```python
df_copy['Date']=pd.to_datetime(df_copy['Date'])
```

```python
df_copy.info()
```

## 4. Feature Engineering and Visualization

#### shot accuracy

```python
df_copy['shot_accuracy']=df_copy['On-Target']/df_copy['Attempts']
```

#### conversion rate

```python
df_copy['conversion_rate']=df_copy['Goal_Scored']/df_copy['Attempts']
```

```python
sum=df_copy.groupby('Team')['Goal_Scored'].sum()
```

```python
sum.plot.bar(figsize=(12,4))
plt.xlabel('Teams')
plt.ylabel('Goal Scored')
plt.title('Conversion')
```

### some additional features
created DataFrames to plot separately later on

```python
#creating a DataFrame of number of goals scored by each team

goals=df_copy.groupby('Team')['Goal_Scored'].sum()
df_goals=pd.DataFrame(goals)
df_goals.head()
```

```python
#creating a DataFrame of number of matches played by each team

matches=df_copy.groupby('Team')['Date'].count()
df_matches=pd.DataFrame(matches)
df_matches.rename(columns={'Date':'Matches_Played'},inplace=True)
df_matches.head()
```

```python
#merge the above 2 DataFrames to form a new: 'df_match'

df_match=pd.merge(df_matches,df_goals,how='outer',on='Team')
df_match.head()
```

```python
#getting goals per match DataFrame

df_match['Goals_Per_Match']= df_match['Goal_Scored']/df_match['Matches_Played']
df_match.head()
```

```python
#getting goals conceded DataFrame

opp_goals=df_copy.groupby('Opponent')['Goal_Scored'].sum()   #groupby and sum
df_oppg=pd.DataFrame(opp_goals)                              #DF
df_oppg.rename(columns={'Goal_Scored':'Goal_Conceded'},inplace=True) #rename col
df_oppg.head()
```

```python
df_oppg.index.names = ['Team'] #rename index
df_oppg.rename(columns={'Opponent':'Team'},inplace=True)  #rename col
df_match=pd.merge(df_match,df_oppg,how='outer',on='Team') #merge
df_match.head()
```

```python
#adding another column: 'Goals Conceded Per Match'

df_match['Goals_Conceded_Per_Match']=df_match['Goal_Conceded']/df_match['Matches_Played']
df_match.head()
```

```python
#x1=list(df_match.index.values)
#index=x1
#storing 
```

```python
df_comp=pd.DataFrame({'Goal_Scored':df_match['Goals_Per_Match'],'Goal_Conceded': df_match['Goals_Conceded_Per_Match']})
ax=df_comp.plot.bar(figsize=(12,4))
plt.xlabel("Teams")
plt.ylabel("Average Goals")
plt.title("Attack vs Defense")
```

### Does "A More Disciplined Team is a Better Defensive Team" ?

```python
df_copy['Discipline']=df_copy['Yellow_Card']+df_copy['Red']+df_copy['Yellow & Red']
df_copy1= df_copy[['Team','Discipline']]
df_copy1.head()
```

```python
df_disc=pd.DataFrame(df_copy.groupby('Opponent')['Discipline'].sum())
df_disc.head()
```

```python
df_disc.index.names=['Team']
```

```python
df_match=pd.merge(df_match,df_disc,how='outer',on='Team')
df_disc.rename(columns={'Opponent':'Team'},inplace=True)
df_match.head()
```

```python
plt.scatter( x=df_match['Discipline'],y= df_match['Goal_Conceded'])
plt.xlabel('Discipline')
plt.ylabel('Goals Conceded')
plt.title('Goal Conceded vs Discipline')
```

- Wanted to establish a relation between how disclipined a team is v/s How poor or good their defense is.
- Though I was not establish a definite trend between the two features but in general, a better disciplined defense means a better defense.
